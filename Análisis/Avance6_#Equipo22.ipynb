**Antecedenctes**

Anteriormente se había observado que el modelo del formato 3, tenía un porcentaje bajo para asegurar que si era el formato 3. 
 
El modelo si realiza bien la extracción de los datos e identificacion del documento, durante las pruebas, siempre extrajo correctamente los datos de cada celda. Contrario a lo que ocurría en los primeros modelos que realizamos porque tenía extraer los datos de formato 4 o formato 3, antes de realizar el modelo diferenciador. 
 
Después de realizar algunas pruebas, nos percatamos que realmente lo que ocasionaba ese bajo porcentaje, se debe a que en los campos que debe extraer la información, tienen mucha varianza en el llenado de cada campo. 
 
Por lo que se sugiere: 
 
1.- Que los nombre de los colaboradores o nombres sean puestos cada uno en un renglón y no en un solo renglón múltiples nombres. 
 
2.- Que los números de WI si se pueden repetir con respecto a los nombres de los empleados porque todos están agrupados en el mismo renglón, así, no se puede determinar si los nombres hacen referencia a una o más personas, y por consecuencia si pertenece a un ítem o a varios items a la vez. Es por eso que se debe de establecer un renglón o fila para cada item o usuario.

# Respuestas a las preguntas
**A. Viabilidad del Modelo**

1. ¿El rendimiento del modelo es lo suficientemente bueno para su implementación en producción?

El rendimiento del modelo actualmente es muy sólido: la clasificación de documentos ofrece resultados de alta fiabilidad y el proceso de extracción de información también cumple con los estándares de exactitud necesarios. Aunque no se observan necesidades inmediatas de mejorar el modelo en sí, sí convendría refinar ciertos aspectos de la operación y del flujo de trabajo. Por ejemplo, asegurar la calidad de los datos desde la etapa de entrenamiento hasta los archivos reales puede mejorar aún más los resultados en producción.

2. ¿Existe margen para mejorar aún más el rendimiento?

Sí, existe margen de mejora. Una de las oportunidades radica en optimizar la generación y nomenclatura de los documentos, lo que reduciría la complejidad en la etapa de clasificación y, por ende, ahorraría recursos. Además, establecer parámetros de formato y espaciado más consistentes, especialmente en tablas, permitiría disminuir errores de entrelazado de filas y facilitaría la extracción automatizada de datos.

3. ¿Cuáles serían las recomendaciones clave para poder implementar la solución?

Realizar un piloto inicial: Implementar la solución en un ambiente controlado para verificar su desempeño y asegurar que se cumplan los requisitos de calidad.
Monitoreo y retroalimentación continua: Aprovechar las funcionalidades de RPA para señalar cuáles filas o campos fueron llenados automáticamente por la IA. De esta manera, los operadores podrán validar la información y ofrecer retroalimentación oportuna para ajustes en el modelo o en el proceso, asegurando un perfeccionamiento constante.
Notificación y comunicación: Asegurarse de que los usuarios finales estén al tanto de la implementación de la solución, así como de mecanismos para interpretar y notificar la correcta operación y mantenimiento del mismo.

4. ¿Qué tareas o procedimientos son accionables para las partes interesadas (stakeholders)?

Comunicación temprana de errores o inconsistencias: Los usuarios deben reportar de inmediato cualquier fallo detectado para poder priorizar mejoras y correcciones de manera ágil.

Monitoreo continuo: Realizar un seguimiento periódico de los registros generados por el RPA y la IA. 

Esto permitirá evaluar la confiabilidad del sistema a lo largo del tiempo y detectar cualquier área que requiera ajustes, garantizando una mejora continua en la precisión y eficiencia de la solución.
